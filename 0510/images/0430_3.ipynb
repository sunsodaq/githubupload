```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = "https://news.naver.com/"
```

```python
page = requests.get(url)
soup = bs(page.text, "html.parser")
```

```python
total_list=[]
elements = soup.select('div')
for i in elements:
    for j in i.find_all('a'):
        for k in j.find_all('span'):
            total_list.append(k.text)
total_list    
```

```python
news_header = soup.find_all('section')
for i in news_header:
    for j in i.find_all('div', {'class':'Ngnb'}):
        for k in j.find_all('a', attrs={"href":"https://www.naver.com"}):
            print(k)
```
